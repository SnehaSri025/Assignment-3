1 What is the difference between a function and a method in python?
- A **function** is a standalone block of code that performs a specific task. It's defined outside of any class and can be called directly by its name.

```python
def greet(name):
  return f"Hello, {name}!"

message = greet("Alice")
print(message)
```

A **method** is a function that is associated with an object (an instance of a class). It's defined inside a class and is called on an object using dot notation. The first parameter of a method is usually `self`, which refers to the instance of the class.

```python
class Dog:
  def __init__(self, name):
    self.name = name

  def bark(self):
    return "Woof!"

my_dog = Dog("Buddy")
sound = my_dog.bark()
print(sound)
```


2 Explain the concept of function arguments and parameters in python?
- **Parameters** are variables defined in a function's definition. They act as placeholders for values that will be passed into the function.

```python
def describe_pet(animal_type, pet_name):
  """Describes a pet."""
  print(f"\nI have a {animal_type}.")
  print(f"My {animal_type}'s name is {pet_name}.")
```

Here, `animal_type` and `pet_name` are parameters.

**Arguments** are the actual values that are passed to a function when it is called.

```python
describe_pet('hamster', 'Harry')
describe_pet(animal_type='dog', pet_name='Lucy')
```

In the first call, `'hamster'` is the argument for `animal_type`, and `'Harry'` is the argument for `pet_name` (positional arguments).

In the second call, `animal_type='dog'` and `pet_name='Lucy'` are keyword arguments



3 What are the different ways to define and call a function in python?
- You can define and call functions in Python in several ways:

**1. Standard Function Definition:
* Use the `def` keyword followed by the function name, parentheses for parameters, a colon, and an indented block of code for the function body.
* Call the function by its name followed by parentheses containing the arguments.

**2. Functions with Default Parameter Values:
* Assign default values to parameters in the function definition. If an argument isn't provided for such a parameter during the call, the default value is used.

**3. Functions with Variable-Length Positional Arguments (`*args`):
* Use `*args` in the parameter list to accept any number of positional arguments. These arguments are collected into a tuple.

**4. Functions with Variable-Length Keyword Arguments (`**kwargs`):
* Use `**kwargs` in the parameter list to accept any number of keyword arguments. These arguments are collected into a dictionary.

**5. Lambda Functions (Anonymous Functions):
* Define small, single-expression anonymous functions using the `lambda` keyword. They can take any number of arguments but can only have one expression.

5 What are iterators in python and how do they differ from iterables?
- In Python:
**Iterators** are objects that allow you to traverse through an iterable. They have two essential methods:

* `__iter__()`: Returns the iterator object itself.
* `__next__()`: Returns the next item from the iterable. When there are no more items, it raises a `StopIteration` exception.

An iterable is something you can loop over, and an iterator is the object that does the actual stepping through the items during the loop. You can think of an iterable as a container, and an iterator as a cursor that moves through the container



6 Explain the concept of generators in Python and how they are defined?
- **Generators** in Python are a simple and powerful way to create iterators. They allow you to declare functions that behave like iterators, meaning they can be used in a `for` loop or with the `next()` function to retrieve items one at a time

There are two main ways to define generators in Python:

**1. Generator Functions:
* Defined like regular functions using the `def` keyword.
* Instead of `return` to send a value and exit, they use `yield` to produce a value and pause execution.
* Can contain multiple `yield` statements.

**2. Generator Expressions:
* A concise way to create anonymous generator functions.
* Similar to list comprehensions but use parentheses `()` instead of square brackets `[]`.
* More memory-efficient than list comprehensions for large datasets as they don't build the entire list in memory.



7  What are the advantages of using generators over regular functions?
- Generators are a more memory-efficient, often faster, and cleaner way to handle sequences, especially large or infinite ones, by producing items lazily and abstracting away the iterator protocol. Regular functions are more suitable when you need to compute and store all results at once in memory.

9 Explain the purpose and usage of the `map()` function in Python?
- The `map()` function in Python is used to apply a given function to each item of an iterable (like a list, tuple, etc.) and return an iterator that yields the results.

**Purpose:**

The primary purpose of `map()` is to **transform** each element of an iterable based on a specified function, without the need for explicit loops. This often leads to more concise and readable code.

**Usage:**

The `map()` function takes two main arguments:

1.  **`function`:** The function to apply to each item of the iterable. This function can be a built-in function, a user-defined function, or a lambda function.
2.  **`iterable`:** The iterable (or iterables) whose items will be processed by the `function`.



10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- The `map()`, `reduce()`, and `filter()` functions in Python are powerful tools for functional programming, each serving a distinct purpose in processing iterables:                

* Use `map()` when you need to apply the same operation to every item in a sequence.
* Use `filter()` when you need to choose certain items from a sequence based on a condition.
* Use `reduce()` when you need to combine all the items in a sequence into a single result.

````python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
class FibonacciIterator:

    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

````

````python
# .Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
  for i in range(n+1):
      yield 2 ** i
````

````python
#.Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filepath):
    with open(filepath, 'r') as file:
      for line in file:
        yield line.strip()

````
